# Dropped Tarots

## Problem statement

Oops! I dropped some tarot cards into a box of pixels. Can you help me retrieve them?

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
X = np.load("dropped_tarots.npy")
```

```python
def visualize_pointcloud(pointcloud, elev=30, azim=30):
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    
    # Extract spatial and color data
    x = pointcloud[:, 0]
    y = pointcloud[:, 1]
    z = pointcloud[:, 2]
    colors = pointcloud[:, 3:6]
    
    scatter = ax.scatter(x, y, z, c=colors, s=5)
    
    # Set axis labels and title
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # Set view angle
    ax.view_init(elev=elev, azim=azim)
    
    # Set equal aspect ratio
    ax.set_box_aspect([1, 1, 1])
    
    plt.tight_layout()
    plt.show()
```

```python
visualize_pointcloud(X)
```
