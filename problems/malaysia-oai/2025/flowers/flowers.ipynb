````python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
````

````python
def visualize_xyzrgb(xyzrgb, stride=1, point_size=1, elev=0, azim=0):
    """
    xyzrgb: np.ndarray of shape (N, 6) with columns x,y,z,r,g,b
    stride: take every k-th point to speed up plotting
    point_size: marker size passed to scatter
    elev, azim: camera angles
    """
    pts = xyzrgb[::stride]

    x = pts[:, 0]
    y = pts[:, 1]
    z = pts[:, 2]
    rgb = pts[:, 3:6]

    # Normalize colors to [0,1] if needed
    if rgb.dtype.kind in "iu" or rgb.max() > 1.0:
        rgb = rgb.astype(np.float32) / 255.0

    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection="3d")

    ax.scatter(x, z, y, s=point_size, c=rgb, depthshade=False)

    ax.view_init(elev=elev, azim=azim)
    ax.set_xlabel("x")
    ax.set_ylabel("z")
    ax.set_zlabel("y")

    # Nice bounds/aspect
    ax.set_xlim(float(x.min()), float(x.max()))
    ax.set_ylim(float(z.min()), float(z.max()))
    ax.set_zlim(float(y.min()), float(y.max() + 1))  # add 1 so the plane is visible
    try:
        ax.set_box_aspect((x.ptp(), y.ptp(), max(1.0, z.ptp())))
    except Exception:
        pass  # older Matplotlib

    # ax.set_axis_off()
    # plt.tight_layout()
    plt.show()
````
